```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
dt= pd.read_csv("D:/adultkmeans.csv")
dt.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
dt['Work Class'] = labelencoder.fit_transform(dt ['Work Class'])
dt.head()
```

```python
dt['marital/status'] = labelencoder.fit_transform(dt ['marital/status'])
```

```python
dt['occupation'] = labelencoder.fit_transform(dt ['occupation'])
dt.head()
```

```python
dt['relationship'] = labelencoder.fit_transform(dt ['relationship'])
dt.head()
```

```python
dt['race'] = labelencoder.fit_transform(dt ['race'])
dt.head()
```

```python
dt['sex'] = labelencoder.fit_transform(dt ['sex'])
dt.head()
```

```python
dt['native-country'] = labelencoder.fit_transform(dt ['native-country'])
dt.head()
dt['class'] = labelencoder.fit_transform(dt ['class'])
dt.head()
```

```python
from sklearn.cluster import KMeans

from sklearn.decomposition import PCA
```

```python
X = dt[['Age','Work Class','Education-num','marital/status','occupation','relationship','race','sex','capital-gain'
        ,'capital-loss','hours-per-week','native-country']]

y = dt[['class']]

print(y)
```

```python
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 0)
y_kmeans = kmeans.fit_predict(X)
print(y_kmeans)

```

```python
from sklearn.metrics import confusion_matrix
j=confusion_matrix(y,y_kmeans)
print(j)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,y_kmeans)
```
